# **Agrupamentos**

### **1-Agrupar por Estado Civil e contar o número de funcionários:**

```python


# Agrupar por Estado Civil e contar o número de funcionários
contagem_estado_civil = dfps.groupBy("Estado civil").agg(count("ID Funcionário").alias("Numero_Funcionarios"))

# Mostrar o resultado
contagem_estado_civil.show()
```

### **2-Agrupar por Departamento e calcular a média de Salário Mensal:**

```python

# Agrupar por Departamento e calcular a média de Salário Mensal, arredondando para 2 casas decimais
media_salario_departamento = dfps.groupBy("Departamento").agg(round(avg("Salário Mensal"), 2).alias("Media_Salario_Mensal"))

# Mostrar o resultado
media_salario_departamento.show()
```

# 3-Agrupar por Departamento e calcular a qtd de treinamento por area

```python
# Agrupar por Departamento e calcular a qtd de treinamento
treinamento_departamento = dfps.groupBy("Departamento").agg(sum("Qtde de treinamentos realizados no último ano").alias("Qtd_Treinamento"))


print(f" Departamento mais treinado ")
# Mostrar o resultado do maior para o menor
treinamento_departamento.orderBy("Qtd_Treinamento", ascending=False).show()
```

### 4-Agrupar genero por departamento, descobri qtd de pessoas por **departamento**

```python
# Agrupar genero por departamento, descobri qtd de pessoas por departamento
departamento_genero = dfps.groupBy("Departamento").agg(count("Gênero").alias("pessoas_por_departamento"))


print(f" Qtd de Pessoas por Departamento ")
# Mostrar o resultado do maior para o menor
departamento_genero.orderBy("pessoas_por_departamento", ascending=False).show()
```
